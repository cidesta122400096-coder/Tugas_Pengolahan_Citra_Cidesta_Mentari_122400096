## **Image Processing (Pengolahan Citra)**

Pada bagian ini dilakukan eksperimen pengolahan citra digital menggunakan library OpenCV. Citra yang digunakan terdiri dari beberapa jenis gambar yaitu object, selfie, dan tekstur.

Tahapan yang dilakukan meliputi:
- Pembacaan citra dari file
- Konversi warna dari BGR ke RGB
- Visualisasi citra menggunakan Matplotlib

Proses konversi warna dilakukan karena OpenCV secara default membaca gambar dalam format BGR, sedangkan Matplotlib menggunakan format RGB, sehingga diperlukan penyesuaian agar hasil tampilan warna sesuai.

---



```python
import cv2
import matplotlib.pyplot as plt

files = ["object.jpg", "selfie.jpg", "tekstur.jpg"]

for f in files:
    img = cv2.imread("assets_ws4/" + f)

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.title(f)
    plt.axis("off")
    plt.show()
```

## **Video Processing (Pemrosesan Video)**

Pada bagian ini dilakukan pemrosesan video menggunakan OpenCV. Video diperlakukan sebagai sekumpulan frame citra yang tersusun secara berurutan dalam domain waktu.

Proses yang dilakukan meliputi:
- Pembacaan file video menggunakan OpenCV
- Ekstraksi video menjadi frame-frame individu
- Konversi setiap frame dari format BGR ke RGB
- Visualisasi frame secara berurutan menggunakan Matplotlib

Konsep ini menunjukkan bahwa video digital tidak lain merupakan kumpulan citra yang ditampilkan secara cepat sehingga menghasilkan ilusi gerakan.

---

```python
import cv2
import matplotlib.pyplot as plt

cap = cv2.VideoCapture("assets_ws4/video_dosen.mp4")

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    if frame_count % 15 == 0:
        plt.imshow(frame_rgb)
        plt.title(f"Frame {frame_count}")
        plt.axis("off")
        plt.show()

    frame_count += 1

cap.release()
```

## **Hasil**

Hasil percobaan menunjukkan bahwa seluruh citra berhasil dibaca dan ditampilkan dengan baik setelah dilakukan konversi warna dari BGR ke RGB. Pada pengolahan citra, tiga jenis gambar yaitu object, selfie, dan tekstur berhasil divisualisasikan dengan benar.

Pada bagian video processing, video berhasil diproses menjadi beberapa frame dan setiap frame dapat ditampilkan secara berurutan. Hal ini menunjukkan bahwa OpenCV mampu menangani data visual dalam bentuk gambar maupun video secara efektif.
